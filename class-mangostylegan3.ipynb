```python
import os
import shutil

src_root = "/kaggle/input/classmango512x512"
dst_root = "/kaggle/working/class_cond"

os.makedirs(dst_root, exist_ok=True)

class_map = {}
for i, class_name in enumerate(sorted(os.listdir(src_root))):
    class_dir = os.path.join(src_root, class_name)
    if os.path.isdir(class_dir):
        label_dir = os.path.join(dst_root, str(i))
        os.makedirs(label_dir, exist_ok=True)
        for j, img_name in enumerate(os.listdir(class_dir)):
            src = os.path.join(class_dir, img_name)
            dst = os.path.join(label_dir, f"{class_name[:3]}_{j}.png")
            shutil.copy(src, dst)
        class_map[class_name] = i

print("✅ Class mapping:", class_map)
```
