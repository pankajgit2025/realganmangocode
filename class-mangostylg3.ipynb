```python
import os
import json
import zipfile

# Path to dataset root folder containing class subfolders
root_dir = "/kaggle/input/classmango512x512/DCCMango_512x512"

# Map class folder names to integer labels
class_names = sorted(os.listdir(root_dir))
label_map = {cls_name: i for i, cls_name in enumerate(class_names)}

print("Class labels:", label_map)

valid_exts = ('.png', '.jpg', '.jpeg', '.bmp')
images = []
labels = {}

# Collect images and labels (relative paths inside the zip)
for class_name in class_names:
    class_path = os.path.join(root_dir, class_name)
    for filename in os.listdir(class_path):
        if filename.lower().endswith(valid_exts):
            rel_path = os.path.join(class_name, filename)
            images.append(os.path.join(class_path, filename))
            labels[rel_path] = [label_map[class_name]]

print(f"Found {len(images)} images.")

# Create dataset.json content
dataset_json = {
    "labels": labels,
    "label_names": class_names
}

# Path to output ZIP file
output_zip = "/kaggle/working/MangoLeaf-cond.zip"

# Create the zip
with zipfile.ZipFile(output_zip, 'w') as z:
    # Write images preserving folder structure inside zip
    for img_path in images:
        rel_path = os.path.relpath(img_path, root_dir)
        z.write(img_path, rel_path)
    # Write dataset.json inside the zip
    z.writestr("dataset.json", json.dumps(dataset_json))

print(f"Created dataset zip: {output_zip}")
```
