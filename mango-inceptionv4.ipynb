```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# Define the input shape and number of classes
input_shape = (299, 299, 3)  # Inception v4 typically uses 299x299 input size
num_classes = 6  # Adjust this based on your dataset

# Load the Inception v4 model with pre-trained weights, excluding the top layers
# Assuming you have a function to load your Inception v4 model
# For demonstration, we will use InceptionV3 as a placeholder
# Replace this with your Inception v4 model loading code
base_model = tf.keras.applications.InceptionV3(weights='imagenet', include_top=False, input_shape=input_shape)

# Freeze the base model layers
for layer in base_model.layers:
    layer.trainable = False

# Add custom layers on top of the base model
x = layers.Flatten()(base_model.output)
x = layers.Dense(256, activation='relu')(x)
x = layers.Dropout(0.5)(x)  # Add dropout to reduce overfitting
x = layers.Dense(num_classes, activation='softmax')(x)  # Output layer for multi-class classification

# Create the new model
model = models.Model(inputs=base_model.input, outputs=x)

# Compile the model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),  # Adjust learning rate
    loss='categorical_crossentropy',  # Multi-class classification
    metrics=['accuracy']
)

# Print the model summary
model.summary()

# Define ImageDataGenerator without augmentation
train_datagen = ImageDataGenerator(rescale=1./255)  # Only rescale for training
val_datagen = ImageDataGenerator(rescale=1./255)    # Only rescale for validation
test_datagen = ImageDataGenerator(rescale=1./255)   # Only rescale for test

# Load Training Data
train_generator = train_datagen.flow_from_directory(
    '/kaggle/input/mangoleafdiseasedataset224x224/MangoLeafDiseaseDataset224x224/train',   # Path to train folder
    target_size=(299, 299), 
    batch_size=32,
    class_mode='categorical'  # Multi-class classification (8 classes)
)

# Load Validation Data
val_generator = val_datagen.flow_from_directory(
    '/kaggle/input/mangoleafdiseasedataset224x224/MangoLeafDiseaseDataset224x224/val',   # Path to validation folder
    target_size=(299, 299),
    batch_size=32,
    class_mode='categorical'
)

# Load Test Data (For final evaluation)
test_generator = test_datagen.flow_from_directory(
    '/kaggle/input/mangoleafdiseasedataset224x224/MangoLeafDiseaseDataset224x224/test',   # Path to test folder
    target_size=(299, 299),
    batch_size=32,
    class_mode='categorical',
    shuffle=False   # No shuffling for test set
)

# Train the Model
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=30,  # Increase the number of epochs for better training
    
)

# Plot Training & Validation Accuracy
plt.figure(figsize=(12, 6))
plt.plot(history.history['accuracy'], label='Train Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='s')

plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training vs. Validation Accuracy  without Augmentation using InceptionV4')
plt.legend()
plt.grid(True)
plt.show()
test_loss, test_acc = model.evaluate(test_generator)
print(f"Test accuracy: {test_acc:.4f}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import seaborn as sns

# Predict on the test set
Y_pred = model.predict(test_generator, steps=test_generator.samples // test_generator.batch_size + 1)
y_pred = np.argmax(Y_pred, axis=1)

# True labels
y_true = test_generator.classes

# Get class labels
class_labels = list(test_generator.class_indices.keys())

# Classification report
print("Classification Report:InceptionV4\n")
print(classification_report(y_true, y_pred, target_names=class_labels))

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_labels, yticklabels=class_labels, cmap='Blues')
plt.title("Confusion Matrix:InceptionV4")
plt.ylabel('Actual Class')
plt.xlabel('Predicted Class')
plt.show()
```

```python
import numpy as np
import os
import matplotlib.pyplot as plt
import tensorflow as tf

# === Predict on Test Set ===
preds = model.predict(test_generator)
y_pred = np.argmax(preds, axis=1)
y_true = test_generator.classes
class_labels = list(test_generator.class_indices.keys())
filenames = test_generator.filenames
test_dir = test_generator.directory

# === Display Correctly Classified Images for Each Class ===
samples_per_class = 5  # how many correct examples to show per class
plt.figure(figsize=(20, 15))

class_count = 0
img_index = 1

for class_id, class_name in enumerate(class_labels):
    # Get correctly classified indices for this class
    correct_indices = np.where((y_true == class_id) & (y_pred == class_id))[0]
    
    if len(correct_indices) == 0:
        continue  # Skip if no correct predictions for this class

    selected = correct_indices[:samples_per_class]  # limit to N examples

    for idx in selected:
        img_path = os.path.join(test_dir, filenames[idx])
        img = tf.keras.preprocessing.image.load_img(img_path, target_size=(299, 299))
        img = tf.keras.preprocessing.image.img_to_array(img) / 255.0

        plt.subplot(len(class_labels), samples_per_class, img_index)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"{class_name}", color='green', fontsize=9)
        img_index += 1

plt.suptitle("✅ Correctly Classified Images per Class", fontsize=18)
plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.show()
```
